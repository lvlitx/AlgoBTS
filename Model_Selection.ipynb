```python
# creating a dictionary with hyper parameters and their values
model_hyperparameters = {
    'log_hype':{
        'C': [1,5,10,20]
    },
    'random_forest_hype':{
        'n_estimators': [10,20,50,100]
    },
    'svc_hype':{
        'kernel': ['linear','rbf','poly','sigmoid'],
        'C': [1,5,10,20]
    },
    'knn_hype':{
        'n_neighbors': [3,5,10]
    }
}

# compare the model with different hyparameter values using gridsearchcv
model_lists = [LogisticRegression(max_iter=1000),RandomForestClassifier(random_state=0),SVC(kernel='linear'),KNeighborsClassifier()]

#Applying the grid search cv on all the mdoel and the hyoper parameters
def ModelSelection(list_of_models, hyperparameters_dictionary):

  result = []

  # Ensure the number of models and hyperparameter sets match
  assert len(list_of_models) == len(hyperparameters_dictionary), "Number of models and hyperparameter sets must be equal"

  for i, model in enumerate(list_of_models):

    key = list(hyperparameters_dictionary.keys())[i] # Get the corresponding key for the hyperparameters

    params = hyperparameters_dictionary[key]

    print(model)
    print(params)
    print('---------------------------------')


    classifier = GridSearchCV(model, params, cv=5)

    # fitting the data to classifier
    classifier.fit(X,Y) # Make sure X and Y are defined in your environment

    result.append({
        'model used' : model,
        'highest score' : classifier.best_score_,
        'best hyperparameters' : classifier.best_params_
    })

  result_dataframe = pd.DataFrame(result, columns = ['model used','highest score','best hyperparameters'])

  return result_dataframe

# Call the function (assuming X and Y are defined)
ModelSelection(model_lists, model_hyperparameters)
```
