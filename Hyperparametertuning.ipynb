```python
import pandas as pd
import numpy as np
from sklearn.metrics import accuracy_score
import sklearn.datasets
from sklearn.svm import SVC
from sklearn.model_selection import GridSearchCV
# from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import RandomizedSearchCV
```

```python
#load the data from the sklearn
breast_cancer_dataset = sklearn.datasets.load_breast_cancer()
```

```python
print(breast_cancer_dataset)
```

```python
#load the data into the data frame
data_frame = pd.DataFrame(breast_cancer_dataset.data, columns=breast_cancer_dataset.feature_names)
```

```python
data_frame.head()
```

```python
# adding the target column to the data frame
data_frame['label'] = breast_cancer_dataset.target
```

```python
data_frame.head()
```

```python
data_frame.shape
```

```python
data_frame.isnull().sum()
```

```python
#Check the distribution of the target variables

data_frame['label'].value_counts()
```

```python
#Seperate the features and the target variables
X = data_frame.drop(columns='label' , axis=1)

Y = data_frame['label']
```

```python
print(X )
```

```python
print(Y
      )
```

```python
# change the values in to a numpy asrray
X = np.asarray(X)

Y = np.asarray(Y)
```

```python
print(X)
# print(Y)
```

```python
model = SVC()
```

```python
# hyperparameters

parameters = {
   'kernel':['linear','poly','rbf','sigmoid'],
   'C':[1,5,10,20]
}
```

```python
classifier = GridSearchCV(model, parameters , cv = 5 )
```

```python
classifier.fit(X,Y)
```

```python
classifier.cv_results_
```

```python
# Find the best parameters
best_parameters = classifier.best_params_
```

```python
print(best_parameters)
```

```python
#highestaccuracy
highest_accuracy = classifier.best_score_
```

```python
print(highest_accuracy)
```

```python
result = pd.DataFrame(classifier.cv_results_)
```

```python
result.head()
```

```python
grid_search_result = result[['param_C','param_kernel','mean_test_score']]
```

```python
grid_search_result
```

```python
import altair as alt

chart = alt.Chart(grid_search_result).mark_point().encode(
    x='param_C',
    y='mean_test_score',
    color='param_kernel',
    tooltip=['param_C', 'param_kernel', 'mean_test_score']
).interactive()

chart
```

```python
#from the above graph the highest accuracy is : 95%
#and the best parameters are :{C:10 , kernel: linear}
```
